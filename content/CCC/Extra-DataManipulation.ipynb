<h3>See the numbers behind the image.</h3>

Run the code below to plot a <em>histagram</em> of the image.<br>
High parts of the graph means that colour shows up a lot, lower parts of the graph means that colour shows up much less.

```python
plt.show(plt.hist(my_Image.ravel(), bins=range(256), fc='k', ec='k'))
```

In the Oxford image,
* There is a spike at <strong>0</strong> - this is the water.
* There is a big spike at about <strong>60</strong> - which is in the healthy vegetation range.
* There is lots of little spikes in the vegetation range.
* There is a final spike at about <strong>250</strong> - which is almost definitely buildings!
