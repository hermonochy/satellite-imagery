<h3>Adding information to your image</h3>

It is not always easy to know what each colour means so, if you want people to understand your image, you can add:
* a title for your image
* a key (called a colour bar)
* a label for the key, to explain what the colours mean

Run the code below to see this in action, then have a go at editing it to describe your image.

```python
my_Image = s.convert( "Nile.jpg" ) # choose an image to work on for this section
#my_Image = s.convert( "Nile.jpg", water=False) # optional - test out this code, what is happening to the water in the image?

plt.imshow(my_Image, cmap='rainbow')
plt.xticks([]) # removes the x-axis values
plt.yticks([]) # removes the y-axis values

plt.colorbar(orientation='horizontal', label="EXPLANATION OF THE COLOURS", ticks=([]))
plt.title('TITLE OF MY IMAGE')
```

Other MatPlotLib functions or additional arguments you may wish to try include:
* Add labels to the axes: <code>plt.xlabel("This is a label for the x axis")</code>
* Remove the colourbar values: <code>plt.colorbar(orientation='horizontal', label="colourbar label", ticks=([]))</code>
* Other label related examples: <a href="https://www.w3schools.com/python/matplotlib_labels.asp" target="_blank">w3schools</a>
